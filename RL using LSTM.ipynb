```python
# !pip install matplotlib
# !pip install pandas
# !pip install keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# Importing the training set
training_set=list()
df=pd.read_csv('data/final_input_file.csv')
df.head()
```

```python
training_set = pd.read_csv('data/final_input_training.csv')
training_set = training_set.iloc[:,1:2].values
# print(training_set)
training_set_size=len(training_set)
print(training_set_size)
predicted=list()
time=list()
```

```python
training_set[:10]
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
training_set = sc.fit_transform(training_set)
```

```python
# Getting the inputs and the ouputs
X1_train = training_set[0:training_set_size]
y_train = training_set[0:training_set_size]
# print(len(X1_train))
# Reshaping
X2_train = np.reshape(X1_train, (training_set_size, 1, 1)) #3 dimesion
# print(X2_train)
```

```python
# define model
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(None, 1)))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint
# specify filepath- this will write a new file for each epoch with the epoch number contained within the filename
filepath="weights/nn_weights-{epoch:02d}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=0, 
                                             save_weights_only=False, save_best_only=False, mode='max')
```

```python
history=model.fit(X2_train, y_train, validation_split=0.3, batch_size = 3, epochs = 300, callbacks=[checkpoint], initial_epoch=0)
```

```python
print(history.history.keys())
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
import time

time_start = time.process_time()

# Predicting the number of jobs for future
test_set = pd.read_csv('data/final_input_testing.csv')
#print(test_set[1])
real_value = test_set.iloc[:,1:2].values


# Getting the predicted jobs of previous values
inputs = real_value
inputs = sc.transform(inputs)
inputs = np.reshape(inputs, (103, 1, 1))
predicted_job = model.predict(inputs)
predicted_job = sc.inverse_transform(predicted_job)
```
