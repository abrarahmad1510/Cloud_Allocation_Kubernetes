```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# Importing the training set
training_set=list()
df=pd.read_csv('data/final_input_file.csv')
df.head()
```

```python
training_set = pd.read_csv('data/final_input_training.csv')
training_set = training_set.iloc[:,1:2].values
print(training_set)
training_set_size=len(training_set)
print(training_set_size)
predicted=list()
time=list()
```

```python
training_set[:10]
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
training_set = sc.fit_transform(training_set)
```

```python
# Getting the inputs and the ouputs
X1_train = training_set[0:training_set_size]
y_train = training_set[0:training_set_size]
# print(len(X1_train))
# Reshaping
X2_train = np.reshape(X1_train, (training_set_size, 1, 1)) #3 dimesion
# print(X2_train)
```

```python
# define model
model = Sequential()
model.add(LSTM(4, activation='sigmoid', input_shape=(None, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
history=model.fit(X2_train, y_train, validation_split=0.3, batch_size = 3, epochs = 300)
```

```python
print(history.history.keys())
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
import time

time_start = time.clock()

# Predicting the number of jobs for future
test_set = pd.read_csv('data/final_input_testing.csv')
#print(test_set[1])
real_value = test_set.iloc[:,1:2].values


# Getting the predicted jobs of previous values
inputs = real_value
inputs = sc.transform(inputs)
inputs = np.reshape(inputs, (103, 1, 1))
predicted_job = model.predict(inputs)
predicted_job = sc.inverse_transform(predicted_job)
```

```python
# Visualising the results
fig=plt.figure(figsize=(12,8))
plt.title('Workload Prediction')
plt.xlabel('Time interval')
plt.ylabel('Number of instances')
plt.plot(test_set.time_interval,predicted_job,color='red',label='predicted', linestyle='dashed')
plt.scatter(df.time_interval,df.no_of_instances,color='blue',label='actual')

plt.legend()
plt.show()
```
